```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(np.arange(-3.0, 3.0, 0.1), np.tanh(np.arange(-3.0, 3.0, 0.1)))
```

```python
points = []
dists = []

for x in np.arange(-3.0, 3.0, 0.1):
    points.append(np.array([x, np.tanh(x)]))
    if len(points) > 2:
        dists.append(np.sqrt(np.sum( (points[-2] - points[-1])**2 )) )
```

```python
plt.plot(dists)
plt.xlabel('x')
plt.ylabel('vzdálenost k bodu x-1')
```

```python
fig = plt.figure(figsize=(18,5))
plt.rc('font', size=15)

plt.subplot(1, 2, 1)
plt.scatter(np.arange(-3.0, 3.0, 0.1), np.tanh(np.arange(-3.0, 3.0, 0.1)))
plt.xlabel('x')
plt.ylabel('tanh(x)')

plt.subplot(1, 2, 2)
plt.scatter(range(len(dists)), dists)
plt.xlabel('x')
plt.ylabel('vzdálenost k bodu x-1')

fig.suptitle('Demonstrace funkce hyperbolický tangens')
 
plt.savefig('../../results/graphs/tanh_demo.pdf')
```
