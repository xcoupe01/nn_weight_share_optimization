```python
class Float8Convertor:
    def __init__(self, mantissa:int=5):

        if mantissa > 6:
            raise Exception('Float8Convertor init error - cannot represent that big mantis')

        self.mantis = mantissa
        self.convert_list = []
        
        num = [False for _ in range(8)]

        print_num(num)
        for _ in range(pow(2,8) - 1):
            num = increment_binary(num)
            print_num(num)
            calculate_float8(num, mantissa)

def increment_binary(num):
    index = 0

    if not num[index]:
        num[0] = True
        return num

    while(index < len(num) and num[index]):
        num[index] = False
        index += 1
    num[index] = True
    
    return num

def print_num(num):
    print([(1 if x else 0) for x in num])


def calculate_float8(num, mantissa):
    print('sign: ' ,num[0])
    print('exponent: ', num[1:len(num) - mantissa])
    print('mantis: ',num[len(num) - mantissa:])

def bit2int(num):
    multiplier = 1
    acc = 0
    for i in range(len(num)):
        acc += multiplier if num[i] else 0
        multiplier *= 2
```
